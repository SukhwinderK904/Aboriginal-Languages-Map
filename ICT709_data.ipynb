```python
import pandas as pd
language = pd.read_csv('aiatsis_lang.csv')
location = pd.read_csv('aiatsis_location.csv')
dictionary = pd.read_csv('aiatsis_dict.csv')
```

```python
language.head()
```

```python
location.head()
```

```python
dictionary.head()
```

```python
language = language[['services','language_code','uri']]
```

```python
language.head(10)
```

```python
language_location = language.merge(location, on=["language_code"], how="outer")
```

```python
language_location.head(10)
```

```python
language_location = language_location[language_location['longitude'] >= 137.99906]
language_location = language_location[language_location['latitude'] >= -28.99909]
```

```python
language_location = language_location.dropna()
```

```python
language_location.head()
```

```python
dictionary_location = dictionary.merge(location, on=["language_code"], how="inner")
```

```python
dictionary_location.head()
```

```python
dictionary_location = dictionary_location[["services","language_code","language_name","indigenous_language","english","pronunciation"]]
```

```python
dicitonary_location = dictionary_location.dropna()
```

```python
dictionary_location.head()
```

```python
language_locationID = dictionary_location.merge(language_location, on=["services", "language_code","language_name"], how="outer")
```

```python
language_locationID.head(100)
```

```python
language_locationID = language_locationID.sort_values("language_name")
```

```python
language_locationID = language_locationID.round(decimals=2)
```

```python
language_locationID = language_locationID.fillna("-")
```

```python
language_locationID.to_csv("language_locationID.csv", encoding="utf-8", index=False)
```
